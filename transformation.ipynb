```python
import pandas as pd
```

```python
orders_df = pd.read_csv("./Data/olist_orders_dataset.csv")
payments_df = pd.read_csv("./Data/olist_order_payments_dataset.csv")
reviews_df = pd.read_csv("./Data/olist_order_reviews_dataset.csv")
items_df = pd.read_csv("./Data/olist_order_items_dataset.csv")
customers_df = pd.read_csv("./Data/olist_customers_dataset.csv")
sellers_df = pd.read_csv("./Data/olist_sellers_dataset.csv")
geolocation_df = pd.read_csv("./Data/olist_geolocation_dataset.csv")
products_df = pd.read_csv("./Data/olist_products_dataset.csv")
products__category_df = pd.read_csv("./Data/product_category_name_translation.csv")
```

```python
orders_df.info()
```

```python
orders_df.drop_duplicates(inplace=True)
orders_df.dropna(how='all', inplace=True)
orders_df["order_purchase_timestamp"] = pd.to_datetime(orders_df["order_purchase_timestamp"])
orders_df["order_delivered_customer_date"] = pd.to_datetime(orders_df["order_delivered_customer_date"])
orders_df["order_estimated_delivery_date"] = pd.to_datetime(orders_df["order_estimated_delivery_date"])
orders_df["actual_delivery_time"] = (orders_df["order_delivered_customer_date"] - orders_df["order_purchase_timestamp"]).dt.days
orders_df["estimated_delivery_time"] = (orders_df["order_estimated_delivery_date"] - orders_df["order_purchase_timestamp"]).dt.days
orders_df["Delay Time"] = orders_df["actual_delivery_time"] - orders_df["estimated_delivery_time"]

orders_df.head()
```

```python
customers_df.info()
```

```python
orders_df.info()
```

```python
final_df = pd.merge(orders_df, customers_df, on="customer_id", how="left")
final_df = pd.merge(final_df, payments_df, on="order_id", how="left")
final_df = pd.merge(final_df, items_df, on="order_id", how="left")
final_df = pd.merge(final_df, products_df, on="product_id", how="left")
final_df = pd.merge(final_df, sellers_df, on="seller_id", how="left")
final_df = pd.merge(final_df, products__category_df, on="product_category_name", how="left")

final_df.info()
```
